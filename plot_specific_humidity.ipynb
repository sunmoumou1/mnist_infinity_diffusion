# 首先绘制一下水汽进行标准化后的概率密度分布图

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the .npy file
file_path = "/home/shepherdchinacan/era5-infinity-diffusion_wo_encoder/process_era5_data_before/specific_humidity_1973_2017.npy"
data = np.load(file_path) * 100

# Replace values less than 0 with 0
data[data < 0] = 0

# data = np.log(data + 1)

# Standardize the data (z-score normalization)
mean = np.mean(data)
std = np.std(data)
standardized_data = (data - mean) / std

# Flatten the data for plotting
flattened_data = standardized_data.flatten()

# Plot the probability density distribution
plt.figure(figsize=(10, 6))
plt.hist(flattened_data, bins=100, density=True, alpha=0.7, color='blue')
plt.title('Probability Density Distribution of Standardized Data')
plt.xlabel('Standardized Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the .npy file
file_path = "/home/shepherdchinacan/era5-infinity-diffusion_wo_encoder/process_era5_data_before/specific_humidity_1973_2017.npy"
data = np.load(file_path) * 100

# Replace values less than 0 with 0
data[data < 0] = 0

data = np.log(1+data)
# Flatten the data for plotting
flattened_data = data.flatten()

# Plot the probability density distribution
plt.figure(figsize=(10, 6))
plt.hist(flattened_data, bins=100, density=True, alpha=0.7, color='blue')
plt.title('Probability Density Distribution of Standardized Data')
plt.xlabel('Standardized Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
